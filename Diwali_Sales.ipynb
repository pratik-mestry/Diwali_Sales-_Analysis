```python
#import python libraries
import numpy as n
import pandas as p
import matplotlib.pyplot as pl
%matplotlib inline
import seaborn as s
```

```python
#import csv file
df=p.read_csv("Diwali Sales Data.csv",encoding="unicode_escape")
```

```python
df.shape
```

```python
df.head(20)
```

```python
df.info()
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.info()
```

```python
p.isnull(df)
```

```python
p.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df["Amount"]=df["Amount"].astype("int")
```

```python
df["Amount"].dtypes
```

```python
df.describe()
```

```python
df[["Age","Orders","Amount"]].describe()
```

```python
#Exploratory Data ANalysis
#Gender
ax=s.countplot(x="Gender",data=df)
```

```python
ax=s.countplot(x="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_gender=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
s.barplot(x="Gender",y="Amount",data=sales_gender)
```

```python
#most of the buyers are female and purchasing power of female is more
```

```python
#Age
df.columns()
```

```python
df.columns
```

```python
ax=s.countplot(data=df,x="Age Group",hue="Gender")
```

```python
ax=s.countplot(data=df,x="Age Group",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
s.barplot(x="Age Group",y="Amount",data=sales_age)
```

```python
#most of the buyers are from the agegroup of 26-35(women)
```

```python
df.columns
```

```python
sales_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
s.set(rc={"figure.figsize":(15,5)})
s.barplot(x="State",y="Orders",data=sales_state)
```

```python
sales_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
s.set(rc={"figure.figsize":(15,5)})
s.barplot(x="State",y="Amount",data=sales_state)
```

```python
#most of orders are from theb uttare pradesh,maharashtra  respectively.
```

```python
#Marital Status
```

```python
ax=s.countplot(data=df,x="Marital_Status")
s.set(rc={"figure.figsize":(8,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
s.set(rc={"figure.figsize":(6,5)})
s.barplot(data=sales_state,x="Marital_Status",y="Amount",hue="Gender")
```

```python
#most of the purchase are married wome
```

```python
#occupation
```

```python
ax=s.countplot(data=df,x="Occupation")
s.set(rc={"figure.figsize":(25,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
s.set(rc={"figure.figsize":(25,5)})
s.barplot(data=sales_state,x="Occupation",y="Amount")
```

```python
#IT sector occupation has more pyurchasing power
```

```python
#Product
```

```python
ax=s.countplot(data=df,x="Product_Category")
s.set(rc={"figure.figsize":(25,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
s.set(rc={"figure.figsize":(25,5)})
s.barplot(data=sales_state,x="Product_Category",y="Amount")
```

```python
#more orders are from the clothing but more amount is from food.
```

```python
sales_state=df.groupby(["Product_ID"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
s.set(rc={"figure.figsize":(25,5)})
s.barplot(data=sales_state,x="Product_ID",y="Amount")
```

```python
#Married women from age group 26-35 years from up,maharashtra and karnataka working in IT,healthcare and avaiationn are more likely to buy the products
#from the food,clothing and electronic category
```
